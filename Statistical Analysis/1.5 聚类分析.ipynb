### Topic: 聚类分析 (Clustering Analysis)简介
聚类分析是一种无监督学习方法，用于将数据分成多个组或簇，使得同一簇内的数据点相似度最大，不同簇之间的相似度最小。我们将学习如何使用Python进行K-means聚类和层次聚类分析，并使用鸢尾花数据集进行示例分析。

### 1. K-means 聚类
K-means聚类是一种常见的聚类算法，通过迭代优化将数据点分配到K个簇中。我们将使用scikit-learn库进行K-means聚类分析。

#### 示例：鸢尾花数据集的K-means聚类

```python
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

# 导入鸢尾花数据集
iris = sns.load_dataset('iris')

# 准备数据
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values

# 创建K-means模型
kmeans = KMeans(n_clusters=3, random_state=0)

# 拟合模型
kmeans.fit(X)

# 预测聚类标签
labels = kmeans.labels_

# 将聚类标签添加到数据集中
iris['cluster'] = labels

# 绘制聚类结果
plt.scatter(iris['sepal_length'], iris['sepal_width'], c=labels, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-means Clustering of Iris Data')
plt.show()
```

### 2. 层次聚类
层次聚类是一种基于树状结构的聚类方法，通过计算数据点之间的距离，逐步合并或分裂簇。我们将使用scipy库进行层次聚类分析。

#### 示例：鸢尾花数据集的层次聚类

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# 计算层次聚类
Z = linkage(X, method='ward')

# 绘制树状图
plt.figure(figsize=(10, 7))
dendrogram(Z, labels=iris.index, leaf_rotation=90, leaf_font_size=5)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

### 示例分析
1. **鸢尾花数据集中有几个聚类？**
   通过K-means聚类结果和层次聚类的树状图，可以判断鸢尾花数据集中有几个聚类。通常情况下，鸢尾花数据集可以分为三个聚类，对应于三种鸢尾花品种。

2. **不同聚类的特征有何异同？**
   通过分析不同聚类的中心和特征分布，可以了解不同聚类的特征异同。例如，不同品种的鸢尾花在萼片长度、萼片宽度、花瓣长度和花瓣宽度上存在显著差异。
