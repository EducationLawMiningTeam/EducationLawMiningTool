### 1.5 基本绘图技术

在Matplotlib和Seaborn中，有许多基础的绘图技术是经常使用的。让我们来看看如何使用这些库创建一些常见的图形。

#### 折线图 (Line Plot)

折线图用于显示数据的变化趋势。

**例子**:

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o')
plt.title('Simple Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

#### 条形图 (Bar Chart)

条形图用于比较不同类别的数据。

**例子**:

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 12, 5, 18]

plt.bar(categories, values)
plt.title('Bar Chart')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

#### 直方图 (Histogram)

直方图用于显示数据的分布情况。

**例子**:

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title('Histogram')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

#### 散点图 (Scatter Plot)

散点图用于显示两个变量之间的关系。

**例子**:

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

让我们创建一个示例条形图，比较四种水果的数量:

```python
import matplotlib.pyplot as plt

# 水果名称和数量
fruits = ['Apple', 'Banana', 'Orange', 'Strawberry']
quantities = [15, 10, 20, 25]

# 创建条形图
plt.bar(fruits, quantities)
plt.title('My favorite fruit quantity comparison')
plt.xlabel('Fruit')
plt.ylabel('Count')
plt.show()
```
